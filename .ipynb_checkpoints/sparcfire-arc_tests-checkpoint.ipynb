```python
# Preamble
from astropy.table import Table, column
import numpy as np
import os

%matplotlib inline
import matplotlib as mpl
from matplotlib import pyplot as plt
# better-looking plots
plt.rcParams['font.family'] = 'serif'
plt.rcParams['figure.figsize'] = (10.0, 8)
plt.rcParams['font.size'] = 18
mpl.ticker.AutoLocator.default_params['nbins'] = 5
mpl.ticker.AutoLocator.default_params['prune'] = 'both'

mpl.rcParams['ps.useafm'] = True
mpl.rcParams['pdf.use14corefonts'] = True
mpl.rcParams['text.usetex'] = True

import sys
sys.path.append('../astro_codes/')
```

```python
samples = Table.read('../fits/SF_fits/samples.fits')
mendel_masses = Table.read('../fits/Simard+Mendel/mendel_masses.fits')
sparcfire = Table.read('../fits/sparcfire/pitch_angles.fits')
```

```python
# Create the sample masks:

in_mass_limit = mendel_masses['in_mass_limit']
in_sparcfire = sparcfire['P'] > 0
in_spiral = np.all([samples['in_spiral'],samples['axial_ok']],
                   axis=0)

sparcfire_spirals = np.all([in_mass_limit,in_sparcfire,in_spiral],
                           axis=0)

print('{} SpArcFiRe spiral galaxies!'.format(sparcfire_spirals.sum()))
```

```python
# Load the individual arc metadata if needed:
L_meta = np.load('../fits/sparcfire/np_metadata/L.npy')
N_meta = np.load('../fits/sparcfire/np_metadata/N_wtd_avg_start.npy')
P_meta = np.load('../fits/sparcfire/np_metadata/P.npy')

P_absolute_meta = np.absolute(P_meta)
P_absolute_meta[P_absolute_meta > 90] = -999
```

#### Following discussion with Steven, we are going to check that our arc _lengths_ and _pitch angles_ are consistent with redshift etc.

```python
# first, create the mask:
N_rows, N_columns = np.shape(P_absolute_meta)
arm_ok = P_absolute_meta > 0

spiral_ok = np.full((N_rows,N_columns),False,dtype='bool')
spiral_ok[sparcfire_spirals] = True

arc_mask = np.all([arm_ok,spiral_ok],axis=0)
```

```python
from contours import kde_contour

fig = plt.figure(figsize=(10,7.5))
ax = plt.subplot(111)

x = L_meta[arc_mask].ravel()
y = P_absolute_meta[arc_mask].ravel()

_ = plt.scatter(x,y,color='k',rasterized=True,s=2,zorder=0)
_ = kde_contour(x,y,fill=True,zorder=1)

ax.set_xlabel('$L$')
ax.set_ylabel('$P$')
ax.set_xlim(40,300)
ax.text(0.95,0.95,'All galaxies ($N_\mathrm{{arcs}}={}$)'.format(arc_mask.sum()),
        ha='right',va='top',transform=ax.transAxes)
```

```python
from bin_statistics import assign_bins
from xy_plots import x_vs_y

assigned_zbins = assign_bins(redshift,N_bins=6,equal_N=True)
bin_edges = [redshift.min()]
for bin_ in np.unique(assigned_zbins):
    in_zbin = assigned_zbins == bin_
    bin_edges.append(redshift[in_zbin].max())
bin_edges = np.array(bin_edges)
    
# Now have the same bin edges for each voronoi binned sample!
```

```python
fig, axarr = plt.subplots(6,2,figsize=(10,15),
                          sharex=True,sharey=False)

for v in np.unique(vbin_assignments):
    in_v = vbin_assignments == v
    P_v = P_arcs[in_v]
    L_v = L_arcs[in_v]
    redshifts_v = redshift_arcs[in_v]
    
    ok_arcs = P_v > 0
    P_v_ok = P_v[ok_arcs].ravel()
    L_v_ok = L_v[ok_arcs].ravel()
    redshift_ok = redshifts_v[ok_arcs].ravel()
    
    ax = axarr[v,0]
    x, y = redshift_ok, P_v_ok
    xy = x_vs_y(x,y)
    _ = xy.show_spearmanr(ax,location='lower right')
    _ = xy.continuous_vs_continuous_binned(ax,bins=8,color=colors[v],lw=2)
    ax.set_ylabel('$P$')
    
    ax = axarr[v,1]
    x, y = redshift_ok, L_v_ok
    xy = x_vs_y(x,y)
    _ = xy.show_spearmanr(ax,location='upper right')
    _ = xy.continuous_vs_continuous_binned(ax,bins=8,color=colors[v],lw=2)
    ax.set_ylabel('$L$')

    
fig.tight_layout()
fig.subplots_adjust(hspace=0)
axarr[-1,0].set_xlabel('redshift')
axarr[-1,1].set_xlabel('redshift')
```
